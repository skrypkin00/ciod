```python
#сортировка Выбором
def selection_sort(arr):  
    # значение i соответствует тому, сколько значений было отсортировано
    for i in range(len(arr)):
        # Мы предполагаем, что первый элемент несортированного сегмента является наименьшим
        lowest_value_index = i
        # Этот цикл перебирает несортированные элементы
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[lowest_value_index]:
                lowest_value_index = j
        # Поменять местами значения самого низкого несортированного элемента с первым несортированным
        arr[i], arr[lowest_value_index] = arr[lowest_value_index], arr[i]
    

a=[6,-30,25,1,-7,8]
for i in range(0,5):
    print(a[i])
print(" ")
selection_sort(a)
for i in range(0,5):
    print(a[i])
```

```python
#Сортировка вставкой
def insertion_sort(arr):
    for i in range(len(arr)):
        element = arr[i]  # Элемент массива
        while arr[i - 1] > element and i > 0:
            arr[i] = arr[i-1]
            i -= 1
        arr[i] = element
    return arr

a=[6,-30,25,1,-7,8]
for i in range(0,5):
    print(a[i])
print(" ")
selection_sort(a)
for i in range(0,5):
    print(a[i])
```

```python
#сортировка Обменом
def bubble(array):
    for i in range(len(array)):
        for j in range(len(array)-i-1):
            if array[j] > array[j+1]:
                array[j],array[j+1] = array[j+1],array[j]
                
a=[6,-30,25,1,-7,8]
for i in range(0,5):
    print(a[i])
print(" ")
selection_sort(a)
for i in range(0,5):
    print(a[i])
```

```python
# Сортировка Шелла
def shell(arr):
    gap = len(arr) // 2 #промежуток   
    while gap>0:
        j=gap
        for i in range(len(arr)-gap):
            if arr[i]>arr[j]:
                arr[i],arr[j]=arr[j],arr[i]
            j+=1
        gap-=1
           
        
    return arr

a=[6,-30,25,1,-7,8]
for i in range(0,5):
    print(a[i])
print(" ")
selection_sort(a)
for i in range(0,5):
    print(a[i])
```

```python
# Быстрая сортировка

def quick_sort(arr, fst=0, lst=None):
    if lst == None:
        lst = len(arr) - 1
        
    if fst >= lst:  #если индекс мен>= бол выходим
        return
 
    i, j = fst, lst
    pivot = arr[(lst + fst) // 2]   #вычисл знач сред индекса
 
    while i <= j:
        while arr[i] < pivot: i += 1
        while arr[j] > pivot: j -= 1   #ищем индекс мен и бол ср значения
        if i <= j:
            arr[i], arr[j] = arr[j], arr[i]
            i, j = i + 1, j - 1
            
    quick_sort(arr, fst, j) #выз функц с послед j
    quick_sort(arr, i, lst) #выз функц с послед i
    
    
a=[6,-30,25,1,-7,8]
for i in range(0,5):
    print(a[i])
print(" ")
selection_sort(a)
for i in range(0,5):
    print(a[i])
```
