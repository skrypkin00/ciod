```python
def BinSearch(arr, x):
    i = 0
    j = len(arr)-1
    while i < j:
        m = int((i+j)/2)
        if x > arr[m]:
            i = m+1
        else:
            j = m
    return j
```

```python
def interpolation(a, k) : 
    left = 0  
    right = len(a) - 1  
    while a[left] < k and k < a[right]:
        mid = int(left + (k - a[left]) * (right - left) / (a[right] - a[left]))  
        if a[mid] < k:
            left = mid + 1
        elif a[mid] > k:
            right = mid - 1
        else:
            return mid

    if a[left] == k:
        return left
    elif a[right] == k:
        return right
    else:
        return left
```
